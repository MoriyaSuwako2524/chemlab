```python
import matplotlib.pyplot as plt
import numpy as np
from Python.Lib.test.test_generators import refleaks_tests
```

```python
class result(object):
    def __init__(self):
        self.total_steps = None
        self.distances = []
        self.energies = []
    def load(self,filename):
        sum = np.load(filename,allow_pickle=True)
        h = sum["history"]
        dis = []
        ene = []
        for i in range(sum["total_steps"]):
            dis.append(h[i]["distance"])
            ene.append(h[i]["energy"])
        idx = np.argsort(dis)
        self.distances = np.array(dis)[idx]
        self.energies  = np.array(ene)[idx]
        self.total_steps = sum["total_steps"]

```

```python
S = result()
S.load("S_summary.npz")
T = result()
T.load("T_summary.npz")
mecp_dis =2.202
mecp_ene = -207.279874
T.energies -= mecp_ene
S.energies -= mecp_ene
```

```python
plt.plot(S.distances,S.energies)
plt.plot(T.distances,T.energies)
plt.scatter(mecp_dis,0)
plt.show()
```
