```python
from ase.io import read, write
from ase.build import make_supercell
import numpy as np

# 读取 unit cell（含 2 个分子）
atoms = read("H2OBPc.mol")

# 构建 3x3x3 超晶胞
P = np.array([[3,0,0],
              [0,3,0],
              [0,0,3]])

supercell = make_supercell(atoms, P)

# 输出为 PDB（给 tleap 用）
write("H2OBPc.pdb", atoms)
```
